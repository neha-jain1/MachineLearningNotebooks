```python
import pandas as pd
```

```python
df = pd.read_csv('movies.csv')
```

```python
df.head()
```

```python
df['desc'] = df['genres'].apply(lambda x : ' '.join(x.split('|')))
```

```python
df.head()
```

```python
df.shape
```

```python
doc = [d for d in list(df['desc'].values)]
```

```python
doc
```

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
```

```python
tfidf = TfidfVectorizer()
```

```python
tfidf_mat = tfidf.fit_transform(doc)
tfidf_mat.shape
```

```python
tfidf_mat
```

## Finding Similarity

```python
from sklearn.metrics.pairwise import cosine_similarity

movies_sim = cosine_similarity(tfidf_mat)
movies_sim = pd.DataFrame(movies_sim)
movies_sim.head()
```

```python
movies_sim.shape
```

## HOw to get movies list

```python
import numpy as np
```

```python
movies_list = df['title'].values
movies_list[:10]
```

```python
# First fetch the id of the movie from movie_list
movie= 'Tom and Huck (1995)'
print(movie)
movie_index = np.where(movies_list==movie)
#print('Movie Index : ', movie_index[0][0])

similar_movies_mat = movies_sim.iloc[movie_index].values

similar_movies_id = np.argsort(-similar_movies_mat)[0][:10]
#print(similar_movies_id)

print(movies_list[similar_movies_id])

```
