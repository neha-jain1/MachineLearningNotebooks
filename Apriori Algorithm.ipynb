```python
import pandas as pd
```

```python
df = pd.read_csv('store_data.csv',header=None)
df.head()
```

```python
df.shape
```

```python
str(df.values[1,3])
```

```python
records  = []
for i in range(0,7501):
    items = []
    for j in range(0,20):
        d = str(df.values[i,j])
        if(d !='nan'):
            items.append(d)
        else:
            break
    records.append(items)
```

```python
records
```

```python
from apyori import apriori
```

```python
patterns = apriori(records, min_support= 0.008, min_confidence=0.10, min_lift=0.5, min_length=1)
```

```python
patterns = list(patterns)
print(len(patterns))
patterns
```
